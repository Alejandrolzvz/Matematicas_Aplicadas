```python

import sklearn
import mglearn
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler
%matplotlib inline
```

```python
pip install mglearn
```

```python
cancer = load_breast_cancer()
cancer.feature_names.shape
print (cancer)
```

```python
list(cancer.feature_names)
```

```python
pca = PCA(n_components = 2) #built 2 components
pca.fit(cancer.data)  #
```

```python
transformada = pca.transform(cancer.data)## reduce total data to a 2 dimensional data
```

```python
print(cancer.data.shape)
print(transformada.shape)
```

```python
mglearn.discrete_scatter(transformada[:,0],transformada[:,1],cancer.target)
plt.legend(cancer.target_names,loc='best')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
```

```python
escala=MinMaxScaler()
escala.fit(cancer.data)
escalada=escala.transform(cancer.data)
pca.fit(escalada)
transformada=pca.transform(escalada)
mglearn.discrete_scatter(transformada[:,0],transformada[:,1],cancer.target)
plt.legend(cancer.target_names,loc='best')
plt.gca()
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
```
